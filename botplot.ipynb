```python
import matplotlib.pyplot as plt
from math import sin, cos, sqrt, pi

def p(x, x1h, x2h, xi):
    return x - x1h * cos(xi) + x2h * sin(xi)

def q(x, x1h, x2h, xi):
    return x - x1h * sin(xi) - x2h * cos(xi)

def create_circle(x, y, r):
    circle= plt.Circle((x,y), radius = r, color='c')
    circle.set_fill(0)
    return circle

def add_shape(patch):
    ax=plt.gca()
    ax.add_patch(patch)

def add_ring(i, x1, x2, xh1, xh2, xi, lmin, lmax):
    x = p(x1[i], xh1[i], xh2[i], xi)
    y = q(x2[i], xh1[i], xh2[i], xi)
    c= create_circle(x,y,lmin)
    add_shape(c)
    c= create_circle(x,y,lmax)
    add_shape(c)
    
sq3 = sqrt(3)

# x1 = [-3, 3, 0]
# x2 = [0, 0, 3 * sq3]
# xh1 = [-1, 1, 0]
# xh2 = [0, 0, sq3]
# lmin = 1
# lmax = 10
# xi = 0

x1 = [-15, 15, 0]
x2 = [-5 * sq3, -5 * sq3, 10 * sq3]
xh1 = [-5, 5, 0]
xh2 = [-5 * sq3 / 3, -5 * sq3 / 3, 10 * sq3 / 3]
lmin = 12
lmax = 27
xi = 2 * pi / 3

add_ring(0, x1, x2, xh1, xh2, xi, lmin, lmax)
add_ring(1, x1, x2, xh1, xh2, xi, lmin, lmax)
add_ring(2, x1, x2, xh1, xh2, xi, lmin, lmax)


plt.axis('scaled')
plt.show()
```
